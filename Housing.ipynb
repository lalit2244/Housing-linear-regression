```python
# Step 1: Import libraries (we are getting our tools)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Step 2: Read the data (like opening a book)
data = pd.read_csv("housing.csv")
print(data.head())

# Step 3: Choose columns (we pick what to use)
X = data[["area"]]  # this is input
y = data["price"]   # this is output (what we want to guess)

# Step 4: Split the data (some to learn, some to test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Step 5: Create and train the model (machine learns from data)
model = LinearRegression()
model.fit(X_train, y_train)

# Step 6: Predict house prices using test data
y_pred = model.predict(X_test)

# Step 7: Evaluate the model (check how good the guesses are)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MAE (Mean Absolute Error):", mae)
print("MSE (Mean Squared Error):", mse)
print("RÂ² Score:", r2)

# Step 8: Draw a picture to show the guesses
plt.scatter(X_test, y_test, color='blue', label='Real Price')
plt.plot(X_test, y_pred, color='red', label='Predicted Price')
plt.xlabel("Area of House")
plt.ylabel("Price of House")
plt.title("House Price Prediction")
plt.legend()
plt.show()
```
